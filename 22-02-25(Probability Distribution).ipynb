**Probability Distribution**


**Implementing Bernoulli Distribution**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
p=0.6
data = bernoulli.rvs(p,size=1000)
plt.bar([0,1],[sum(data == 0),sum(data == 1)],
        color=['yellow','lavender'],tick_label=["Failure (0)","Success (1)"])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Bernoulli Distribution (p=0.6)")
plt.show()
```

**Implementing Normal Distribution**

```python
from scipy.stats import norm
mu=70
sigma=10
x=np.linspace(mu - 4*sigma,mu+4*sigma,100)
pdf = norm.pdf(x,mu,sigma)
plt.plot(x,pdf,color='orange',label=f"N({mu},{sigma}{sigma})")
plt.xlabel("Exam Scores")
plt.ylabel("Probability Density")
plt.title(f"Normal Distribution (={mu},={sigma})")
plt.legend()
plt.grid()
plt.show()
```

```python
from scipy.stats import norm
z=np.linspace(-4,4,1000)
pdf=norm.pdf(z,0,1)
plt.plot(z,pdf,color='lavender',
         label="Standard Normal Distribution (=0,=1)")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.title("Standard Normal Distribution (Z-Distribution)")
plt.axvline(x=0,color='red',linestyle='--',label="Mean (Z=0)")
plt.legend()
plt.grid()
plt.show()
```

```python
import seaborn as sns
#set random seed for 
np.random.seed(42)
income_data = np.random.exponential(scale=50,size=1000)
plt.figure(figsize=(8,4))
sns.histplot(income_data,bins=50,kde=True,color='y')
plt.title("Original Incoem Distribution (skewed)")
plt.xlabel("Income ($1000)")
plt.ylabel("Frequency")
plt.show()
# perform sampling and calculate means
sample_means=[]
n_samples=1000 # Number of samples
sample_size=30 # sample size
for _  in range(n_samples):
    sample=np.random.choice(income_data,size=sample_size)
    sample_means.append(np.mean(sample))
plt.figure(figsize=(8,4))
sns.histplot(sample_means,bins=50,kde=True,color='y')
plt.title(f"Sampling Distribution of sample Mean(n={sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
