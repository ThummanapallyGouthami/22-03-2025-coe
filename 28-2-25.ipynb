***Scatter Plot***

***Histogram***

```python
# Histogram for Height
plt.figure(figsize=(10, 5))
sns.histplot(df['Height (cm)'], kde=True, bins=10,color='lavender')
plt.title('Distribution of Height')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.show()

# Histogram for Weight
plt.figure(figsize=(10, 5))
sns.histplot(df['Weight (kg)'], kde=True, bins=10,color='pink')
plt.title('Distribution of Weight')
plt.xlabel('Weight (kg)')
plt.ylabel('Frequency')
plt.show()
```

***Checking Independent and Dependent Variables***

```python

## Independent features
x= df['Weight (kg)']
type(x)
```

```python
#checking the type of dependent variable
x= df[['Weight (kg)']]
type(x)
```

***Independent Feature should be in the form of DataFrame or 2D Array***

```python
np.array(x).shape
```

***Dependent Feature should be in series or 1D-Array***

```python
#checking the type of dependent variable
y= df['Height (cm)']
type(y)
```

***Train,Test,Split***

```python
#Train,Test,Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

***fit_transform***


-> fit_transform is used for train data for test data we use transform
 

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit_transform(x_train)
```

```python

x_test = scaler.transform(x_test)
x_test 

```

***Linear Regression***

```python
from sklearn.linear_model import LinearRegression

regression = LinearRegression(n_jobs=-1)
regression.fit(x, y)
print("Intercept:", regression.intercept_)
print("Coefficient:", regression.coef_)
```

```python
## plot Training data plot best fit line
scatter = plt.scatter(x_train,y_train)
plt.show(scatter)

slope_of_best_fit = plt.plot(x_train,regression.predict(x_train))

plt.show(slope_of_best_fit) 

plt.show(slope_of_best_fit)


```
